```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Step 1: Create the dataset
np.random.seed(42)
data = {
    'ID': np.arange(1, 501),
    'TV': np.random.randint(50, 300, 500),
    'Radio': np.random.randint(5, 100, 500),
    'Newspaper': np.random.randint(1, 50, 500),
    'Sales': np.random.randint(100, 1000, 500)  # Target variable
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Step 2: Define independent (X) and target (y) variables
X = df[['TV', 'Radio', 'Newspaper']]
y = df['Sales']

# Step 3: Split the dataset into training (70%) and testing (30%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Train a simple linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 5: Print training and testing sets
print("Training Set (First 5 rows):")
print(X_train.head())
print(y_train.head())

print("\nTesting Set (First 5 rows):")
print(X_test.head())
print(y_test.head())

# Step 6: Print model coefficients and intercept
print("\nModel Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

np.random.seed(42)
data = {
    'ID' : np.arange(1, 501),
    'TV' : np.random.randint(1, 20, 500),
    'Radio' : np.random.randint(5, 200, 500),
    'NewsPaper' : np.random.randint(2, 300, 500),
    'Selas' : np.random.randint(10, 50, 500)      #Target Veriable      
}

dataset = pd.DataFrame(data)

X = dataset[['TV', 'Radio', 'NewsPaper']]
y = dataset['Selas']

# Trainning and testing of data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

# Linear Regression
model = LinearRegression()
model.fit(X_train, y_train)

# print models
print('\n Trainning set')
print(X_train.head())
print(y_train.head())

print('\n Testing set')
print(X_test.head())
print(y_test.head())

# coeficient and intercept
print("Model Coeficient",model.coef_)
print("Model Intercept",model.intercept_)
```
